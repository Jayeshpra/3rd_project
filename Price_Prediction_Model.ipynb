```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
pd.set_option('expand_frame_repr',False)
```

```python
df = pd.read_csv('Sep-Potato.csv', header=1)
df['Price Date'] = pd.to_datetime(df['Price Date'], format='%d-%b-%y')
df_cleaned = df.drop_duplicates(subset=['Price Date'])
df_sorted = df_cleaned.sort_values(by='Price Date')
Extracted_features = df_sorted[['District Name', 'Commodity', 'Min Price (Rs./Quintal)', 'Max Price (Rs./Quintal)', 'Modal Price (Rs./Quintal)', 'Price Date']]
Extracted_features.set_index('Price Date', inplace=True)
full_date = pd.date_range(start='2024-09-01', end='2024-09-30')
full_Extracted_features = Extracted_features.reindex(full_date)
full_Extracted_features['Date'] = full_date
full_Extracted_features['Max Price (Rs./Quintal)'] = full_Extracted_features['Max Price (Rs./Quintal)'].interpolate(method='linear')
full_Extracted_features['Min Price (Rs./Quintal)'] = full_Extracted_features['Min Price (Rs./Quintal)'].interpolate(method='linear')
full_Extracted_features['Modal Price (Rs./Quintal)'] = full_Extracted_features['Modal Price (Rs./Quintal)'].interpolate(method='linear')
full_Extracted_features['Max Price (Rs./Quintal)'] = full_Extracted_features['Max Price (Rs./Quintal)'].round(0)
full_Extracted_features.iloc[0, 2:5] = full_Extracted_features.iloc[1:3, 2:5].mean()
full_Extracted_features['District Name'], full_Extracted_features['Commodity'] = 'Ahmedabad', 'Potato'
display(full_Extracted_features)
```

```python
df3 = pd.read_csv('Sep-bottle gourd.csv', header=1)
df3['Price Date'] = pd.to_datetime(df3['Price Date'], format='%d-%b-%y')
df3_cleaned = df3.drop_duplicates(subset=['Price Date'])
df3_sorted =  df3_cleaned.sort_values(by='Price Date')
Extracted_features3 = df3_sorted[['District Name', 'Commodity', 'Min Price (Rs./Quintal)', 'Max Price (Rs./Quintal)', 'Modal Price (Rs./Quintal)', 'Price Date']]
Extracted_features3.set_index('Price Date', inplace=True)
full_date3 = pd.date_range(start='2024-09-01', end='2024-09-30')
full_Extracted_features3 = Extracted_features3.reindex(full_date3)
full_Extracted_features3['Date'] = full_date3
full_Extracted_features3['Max Price (Rs./Quintal)'] = full_Extracted_features3['Max Price (Rs./Quintal)'].interpolate(method='linear')
full_Extracted_features3['Min Price (Rs./Quintal)'] = full_Extracted_features3['Min Price (Rs./Quintal)'].interpolate(method='linear')
full_Extracted_features3['Modal Price (Rs./Quintal)'] = full_Extracted_features3['Modal Price (Rs./Quintal)'].interpolate(method='linear')
full_Extracted_features3.iloc[0, 2:5] = full_Extracted_features.iloc[1:3, 2:5].mean()
full_Extracted_features3['District Name'], full_Extracted_features3['Commodity'] = 'Ahmedabad', 'Bottle gourd'
display(full_Extracted_features3)
```

```python
# Weather conditin for Ahmedabad
weather = pd.read_csv('Sep-Weather-Ahmedabad.csv')
selected_features = weather[['datetime', 'tempmax', 'tempmin', 'temp', 'dew', 'humidity', 'windspeed', 'sealevelpressure']]
selected_features['Date'] = selected_features['datetime']
selected_features.set_index('datetime', inplace=True)
selected_features['Date'] = pd.to_datetime(selected_features['Date'])
display(selected_features)
```

```python
Potato = full_Extracted_features.merge(selected_features, on='Date')
display(Potato)
Onion = full_Extracted_features1.merge(selected_features, on='Date')
display(Onion)
Cabbage = full_Extracted_features2.merge(selected_features, on='Date')
display(Cabbage)
Bottle_gourd  = full_Extracted_features3.merge(selected_features, on='Date')
display(Bottle_gourd)
```

```python
X = Potato[['tempmax', 'tempmin', 'temp', 'dew', 'humidity', 'windspeed', 'sealevelpressure']]
Y = Potato['Max Price (Rs./Quintal)']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42) 

scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(x_train_scaled, y_train)
y_pred = model.predict(x_test_scaled)
predicted_price = pd.DataFrame(y_pred)
display("Predicted Prices", predicted_price)
```

```python
plt.scatter(y_test, y_pred)
plt.xlabel('Actual Max Price')
plt.ylabel('Predicted Max Price')
plt.title('Actual vs Predicted Max Price')
plt.show()
```
